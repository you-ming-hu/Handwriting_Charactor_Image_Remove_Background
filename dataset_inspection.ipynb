```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib
pd.options.display.max_rows = 800
```

```python
#training data存放的路徑
image_dict = {}
for image in pathlib.Path(r'D:\DL_competition_Dataset\handwritten_competition\backup\train').iterdir():
    character = image.with_suffix('').name[-1]
    image_dict.setdefault(character,[]).append(image)
```

```python
char_freq_table = pd.Series(data = {k:len(v) for k,v in image_dict.items()},name='num_imgs').to_frame().sort_values('num_imgs',ascending=False)
```

```python
char_freq_table.reset_index()
```

```python
#因為plt預設無法處理中文，所以先drop中文index
char_freq_img = char_freq_table.reset_index(drop=True).plot.bar(figsize=(40,10))
char_freq_img
plt.show()
```
